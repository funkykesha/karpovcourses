# 1. Сколько студентов успешно сдали только один курс?

##  Успешная сдача — это зачёт по курсу на экзамене

```python
base_describe(assessments)
```

```python
# ищем экзамены
assessments_exam = assessments.query('assessment_type == "Exam"')
assessments_exam.head()
```

```python
# отбираем в результатах тестов студентов только экзамены и добавим продолжительность курсов
studentAssessment_exam = studentAssessment.merge(assessments_exam, on='id_assessment', how='inner') \
                                          .merge(courses, on=['code_presentation','code_module'], how='left')
studentAssessment_exam.head()
```

```python
# проверяем есть ли перезачеты
studentAssessment_exam.query('is_banked == 1')
```

Перезачетов нет.

```python
# добавляем колонку с результатом - Сдал / Не сдал
studentAssessment_exam['result'] = studentAssessment_exam.score.apply(lambda x: 'passed' if x >= 40 else 'not_passed')
studentAssessment_exam.head()
```

```python
# преобразуем в двоичный формат
studentAssessment_exam_dummies = pd.get_dummies(studentAssessment_exam['result'])
studentAssessment_exam_dummies.head()
```

```python
# склеиваем с данными
exam = pd.concat([studentAssessment_exam, studentAssessment_exam_dummies], axis=1)
exam.head()
```

```python
# заполянем нули
exam = exam.fillna(0)
exam.head()
```

```python
# все кто сдал экз позже дедлайна
exam.query('date != 0 & date_submitted > date')
```

```python
# все кто сдал экз позже окончания курса
exam.query('date_submitted > module_presentation_length')
```

У большинства стоят балы, значит предполагаем что они сдали.
Значит дата не влияет

```python
# оставляем только успешно сдавших студентов
passes_exam = exam.query('result == "passed"')
passes_exam.head()
```

```python
# считаем сколько успешных сдач у каждого студента, фильтруем по количеству (== 1) и считаем сколько таких студентов
passes_exam.groupby('id_student', as_index=False)['score'].count().query('score == 1')['score'].sum()
```

## Ответ: 3802 человека сдало только один курс

# 4. Выяви самые популярные курсы (ТОП-3) по количеству регистраций на них. А также курсы с самым большим оттоком (ТОП-3). 

```python
# заполняем NaN
studentRegistration.head()
```

```python
# добавляем столбец с обозначением отменил ли обучение студент
studentRegistration['leave_course'] = studentRegistration.date_unregistration.notna()
studentRegistration.head()
```

```python
# Считаем количество регистраций и отмен и сразу переименуем столбец
course_reg = studentRegistration.groupby(['code_module','code_presentation'], as_index=False) \
                                .agg({'leave_course':'sum','id_student':'count'}) \
                                .rename(columns={'id_student':'sum_reg'})

course_reg.head()
```

```python
# Добавляем столбцы с конверсией в отмену регистрации, 
# а также разницу между всеми регистрациями и отменами регистраций (количественный отток)
course_reg['CR'] = course_reg.leave_course / course_reg.sum_reg
course_reg.head()
```

```python
# ТОП 3 курса по количеству регистраций
course_reg.sort_values('sum_reg', ascending=False).head(3)[['code_module','code_presentation','sum_reg']]
```

```python
# ТОП 3 курса с самым большим оттоком в % выражении
course_reg.sort_values('CR', ascending=False).head(3)[['code_module','code_presentation','CR']]
```

```python
# ТОП 3 курса с самым большим оттоком в количественном выражении
course_reg.sort_values('leave_course', ascending=False).head(3)[['code_module','code_presentation','leave_course']]
```

# 5. Напиши функцию на python, позволяющую строить когортный (семестровый) анализ

## В период с начала 2013 по конец 2014 выяви семестр с самой низкой завершаемостью курсов и самыми долгими средними сроками сдачи курсов.

```python
def kogort_presentation(assessments, studentAssessment):
    # ищем экзамены
    assessments_exam = assessments.query('assessment_type == "Exam"')
    
    # отбираем в результатах тестов студентов только экзамены
    studentAssessment_exam = studentAssessment.merge(assessments_exam, on='id_assessment', how='inner')
    
    # добавляем колонку с результатом - Сдал / Не сдал
    studentAssessment_exam['result'] = studentAssessment_exam.score \
                                        .apply(lambda x: 'passed' if x >= 40 else 'not_passed')
    
    # преобразуем в двоичный формат
    studentAssessment_exam_dummies = pd.get_dummies(studentAssessment_exam['result'])
    
    # склеиваем с данными
    exam = pd.concat([studentAssessment_exam, studentAssessment_exam_dummies], axis=1)
    
    # Семестры с количеством попыток сдать экзамен, количеством успехов и провалов
    presentation_result = exam \
                                    .groupby('code_presentation', as_index=False) \
                                    .agg({'score': 'count', \
                                          'passed': 'sum'})

    # Семестры со средним сроком сдачи успешных экзаменов
    avg = exam.query('result == "passed"').groupby('code_presentation', as_index=False) \
                                          .agg({'date_submitted': 'mean'})
    avg['date_submitted'] = round(avg['date_submitted'], 2)
    
    # Склеиваем
    presentation_result = presentation_result.merge(avg, on='code_presentation', how='inner')
    
    # Переименовываем столбцы для убоства чтения
    presentation_result = presentation_result.rename(columns={'score': 'attempts', 'date_submitted': 'avg_date'})
    
    # Вычисляем конверсию сдачи и переводим в вид %
    presentation_result['CR'] = round(presentation_result['passed'] / presentation_result['attempts'] * 100, 2)
    
    # Удаляем лишние столбцы
    presentation_result = presentation_result.drop(columns=['attempts','passed'])
    
    # Вычисляем семестр с самой низкой завершаемостью курсов
    low_cr = presentation_result.sort_values('CR').head(1)['code_presentation']
    
    # Вычисляем семестр с самым долгим средним сроком сдачи курса
    low_avg = presentation_result.sort_values('avg_date').head(1)['code_presentation']
    
    if low_cr[0] == low_avg[0]:
        print('Семестр с самой низкой завершаемостью курсов и самым долгим средним сроком сдачи курса -', low_cr[0])
    else:
        print('Разные семестры с низкими показателями')
        print('Семестр с самой низкой завершаемостью курсов -', low_cr[0])
        print('Семестр с самым долгим средним сроком сдачи курса -', low_avg[0])
```

```python
kogort_presentation(assessments, studentAssessment)
```

# 6. Построить адаптированные RFM-кластеры студентов, чтобы качественно оценить свою аудиторию. 

Метрики:
- R - среднее время сдачи одного экзамена
- F - завершаемость курсов
- M - среднее количество баллов, получаемое за экзамен. 

Описать как создавались кластеры. 

Для каждого RFM-сегмента построить границы метрик recency, frequency и monetary для интерпретации этих кластеров. 

###### Пример: 

RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 0.1 до 0.5, monetary от 55 до 72 баллов. 

```python
exam.head()
```

Во всех рассчетах будем брать средние значения для студента по всем попыткам сдать экзамен

Вычислим за сколько дней до конца курса сдают экзамен - понадобится для recency (среднее время сдачи одного экзамена)

```python
exam['days_to_deadline'] = exam['module_presentation_length'] - exam['date_submitted']
exam.head()
```

Возьмем CR из таблицы Assessment_exam - понадобится для frequency (завершаемость курсов)

```python
Assessment_exam
```

Добавим в таблицу со всеми студентами "exam"

```python
exam = exam.merge(Assessment_exam[['id_assessment','CR']], on='id_assessment', how='left')
exam.head()
```

Для оценки monetary (среднее количество баллов, получаемое за экзамен) будем использовать "score"

Теперь посчитаем среднее нужных метрик для каждого студента

```python
exam_student = exam.groupby('id_student', as_index=False).agg({'days_to_deadline': 'mean', \
                                                'CR': 'mean', \
                                                'score': 'mean'})
exam_student.head()
```

Оценим распределение "days_to_deadline"

```python
sns.distplot(exam_student.days_to_deadline, bins=30)
```

В распределении явно 4 пика.

Попробуем поделить на квартили всех кто успел и всех кто опаздал сделаем 5ой категорией

```python
pd.qcut(exam_student.query('days_to_deadline >= 0')['days_to_deadline'], q=4).value_counts()
```

Получается, что сдают либо за 3 дня до окончания курса либо опаздывают

Зададим границы из квартилей, взяв промежутки и установив недостижимые границы (возьмем максимальную длину семестра). 

Таким образом получим 5 границ, где ниже самого наименьшего все что сдавал позже чем за три дня до окончания курса. В качестве границы для самого низкого возьмем самую позднюю дату сдачи и (отнимем еще 5 дней) от окончания курса.

```python
exam_student.days_to_deadline.min()
```

```python
exam.module_presentation_length.max()
```

```python
cut_bins = [-269, 3, 11, 19, 24, 269]
bins_labels = [5, 4, 3, 2, 1]
exam_student['recency'] = pd.cut(exam_student['days_to_deadline'], bins=cut_bins, labels=bins_labels)
exam_student.head()
```

Теперь оценим CR.

```python
sns.distplot(exam_student.CR, bins=30)
```

Есть выраженные 5 пиков, поэтому разделим всех на 5 частей

```python
 pd.qcut(exam_student['CR'], q=5).value_counts()
```

```python
bins_labels = [5,4,3,2,1]
exam_student['frequency'] = pd.qcut(exam_student['CR'], q=5, labels=bins_labels)
exam_student
```

Оцениваем score

Всех у кого среднее меньше 40 отнесем в отдельную самую худшую категорию и оценим оставшихся студентов.

```python
sns.distplot(exam_student.query('score >= 40').score, bins=30, kde=False)
```

Просматривается 4 границы, попробуем поделить на 4 части.

```python
pd.qcut(exam_student.query('score >= 40')['score'], q=4).value_counts()
```

Возьмем промежутки и добавим крайние недостижимые границы -0.1 и 100.1 (меньше нуля нельзя набрать, как и набрать больше 100)

```python
cut_bins = [-0.1, 39.999, 56, 69, 84, 100.1]
bins_labels = [5, 4, 3, 2, 1]
exam_student['monetary'] = pd.cut(exam_student['score'], bins=cut_bins, labels=bins_labels)
exam_student.head()
```

И создадим столбец RFM

```python
exam_student['RFM'] = exam_student.recency.astype(str) + exam_student.frequency.astype(str) \
                                                                                + exam_student.monetary.astype(str)
exam_student['RFM'] = exam_student['RFM'].str.rstrip('.0')
exam_student.sort_values('RFM')
```
