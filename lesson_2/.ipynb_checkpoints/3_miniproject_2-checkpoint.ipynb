```python
# Переименовываем столбцы
```

```python
bookings.columns = bookings.columns \
    .str.lower() \
    .str.replace(' ', '_')
```

```python
# Из какой страны совершали больше всего бронирований
```

```python
bookings.query('is_canceled == 0') \
    .groupby(['country'], as_index = False) \
    .agg({'reservation_status_date' : 'count'}) \
    .sort_values(['reservation_status_date'], ascending= False) \
    .head()
```

```python
# На сколько ночей бронируют отели типа City Hotel? Resort Hotel?
```

```python
stays_total_nights = bookings.query('hotel in ("City Hotel", "Resort Hotel")') \
    .groupby('hotel') \
    .agg({'stays_total_nights' : 'mean'}) \
    .round(2)
stays_total_nights
```

```python
# Количество изменений типа бронирования (включая отмену)
```

```python
bookings.query('assigned_room_type != reserved_room_type').agg({'lead_time' : 'count'})
```

```python
# Сравнение самых часто-бронируемых месяцев в 2016 и 2017
```

```python
# Сначала ищем самый популярный месяц в 2016
```

```python
bookings.query('arrival_date_year in ("2016")') \
    .query('is_canceled == 0') \
    .groupby(['arrival_date_year', 'arrival_date_month']) \
    .agg({'hotel' : 'count'}) \
    .sort_values(['hotel'], ascending = False)
```

```python
# Так как это октябрь, то сравниваем с 2017
```

```python
bookings.query('arrival_date_year in ("2017")') \
    .query('is_canceled == 0') \
    .groupby(['arrival_date_month'], as_index = False) \
    .agg({'hotel' : 'count'}) \
    .sort_values(['hotel'], ascending = False)
```

```python
# В каких месяцах среди в каждом из годов 2015-2017 тип отелей City Hotel отменялся чаще всего?
```

```python
bookings.query('arrival_date_year in ("2015", "2016", "2017")') \
        .query('is_canceled == 1') \
        .query('hotel in ("City Hotel", "Resort Hotel")') \
        .groupby(['arrival_date_year', 'arrival_date_month'], as_index = False) \
        .agg({'hotel' : 'count'}) \
        .sort_values(['hotel'], ascending = False)
```

```python
# Укого из adults, children и babies большее среднее значение?
```

```python
bookings.agg({'adults' : 'mean', 'children' : 'mean', 'babies' : 'mean'})
```

```python
# Для отелей какого типа среднее значение количества детей (всех) наибольшее?
```

```python
# Создаем столбец со всеми детьми
```

```python
bookings['total_kids'] = bookings['children'] + bookings['babies']
bookings
```

```python
# Считаем среднее для всех детей по каждому типу отелей
```

```python
bookings.groupby('hotel', as_index = False) \
        .agg({'total_kids' : 'mean'}) \
        .round(2)
```

```python
# Посчитать сколько бронирований потеряно из-за начиличия детей
```

```python
# Создаем переменную наличия детей
```

```python
bookings['has_kids'] = bookings.total_kids != 0.0

bookings.groupby('has_kids').agg({'total_kids' : 'count'})
```

```python
# Считаем потерянные бронирования с детьми
```

```python
a = bookings.query('has_kids == True and is_canceled == 1').agg({'has_kids' : 'count'})
b = bookings.query('has_kids == True').agg({'has_kids' : 'count'})
c = a / b * 100
c = c.round(2)
print(a, b, c, sep='\n')
```

```python
# Считаем потерянные бронирования без детей
```

```python
a = bookings.query('has_kids == False and is_canceled == 1').agg({'has_kids' : 'count'})
b = bookings.query('has_kids == False').agg({'has_kids' : 'count'})
c = a / b * 100
c = c.round(2)
print(a, b, c, sep='\n')
```

```python
# Считаем потерянные бронирования в целом
```

```python
a = bookings.query('is_canceled == 1').agg({'has_kids' : 'count'})
b = bookings.agg({'has_kids' : 'count'})
c = a / b * 100
c = c.round(2)
print(a, b, c, sep='\n')
```
