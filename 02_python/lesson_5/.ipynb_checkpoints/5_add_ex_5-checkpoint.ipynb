```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# В папке subsid (shared/homeworks/python_ds_miniprojects/5_subsid) находятся файлы (tm_sales_1, tm_sales_2, ...)
# продажами продуктов через телемаркетинг
# Каждый файл содержит, как минимум, 4 колонки (поля): FILIAL_ID, SUBS_ID, PROD_ID, ACT_DTTM.
# Суть задачи в том, чтобы проверить подключения продуктов определенным пользователям
# соединив файлы о продажах с логами по подключениям в системе.

# Особенности данных:

# сотрудники телемаркетинга не всегда указывают полный id, если 'id' нет в начале SUBS_ID, то нужно его добавить
# поля в файлах могут быть расположены абсолютно случайным образом, но названия полей статичны
# продажа не засчитывается, если отключение (END_DTTM) произошло меньше чем через 5 минут после подключения (START_DTTM)
# если в файле с продажами встречается строка без указанного SUBS_ID, она пропускается
# Сохраните результат в датафрэйм с разделителем ;, содержащий корректные подключения.

# Note: обратите внимание на то, как pandas переводит дату из строки, возможно вам понадобится параметр format
# Пример содержания итогового файла (колонка difference это разница между START_DTTM и END_DTTM):

# ACT_DTTM;FILIAL_ID;difference
# 20-03-2020 14:59;1;255 days 09:01:00
```

```python
path = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-a-gajbadulin-24/\
shared/homeworks/python_ds_miniprojects/5_subsid/subsid'
```

```python
data_row = []
for dir_1 in os.listdir(path):
    if dir_1 == 'prod_activations_logs.csv':
        continue
    local_data = pd.read_csv(path + '/' + dir_1, sep=';')
    print(local_data)

data_row
```

```python
data = pd.DataFrame()
for dir_1 in os.listdir(path):
    if dir_1 == 'prod_activations_logs.csv':
        continue
    local_data = pd.DataFrame()
    local_data = local_data.append(pd.read_csv(path + '/' + dir_1, sep=';', parse_dates=['ACT_DTTM']))
    data = data.append(local_data)

data = data.sort_values('FILIAL_ID')

# Удаляем нулвевые значения
data = data.dropna(subset=['SUBS_ID']) 

# Добавляем id если нет
data['SUBS_ID'] = data['SUBS_ID'].apply(lambda x: x if x.startswith('id') == True else 'id' + x)



data 
```

```python
path_2 = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-a-gajbadulin-24/shared/homeworks/python_ds_miniprojects/5_subsid/subsid/prod_activations_logs.csv'
df_prod = pd.read_csv(path_2, sep=';').reset_index().rename(columns={'index': 'FILIAL_ID'})
                        
df_prod
```

```python
# Джойним
```

```python
data = data.merge(df_prod, how='inner', on=['SUBS_ID', 'PROD_ID', 'FILIAL_ID'])
```

```python
data
```

```python
# Если отключение произошло меньше чем за 5 минут, то удаляем заказ
```

```python
data['ACT_DTTM'] = pd.to_datetime(data['ACT_DTTM'], format='%Y-%m-%d %H:%M:%S')
data['START_DTTM'] = pd.to_datetime(data['START_DTTM'], format='%d-%m-%Y %H:%M')
data['END_DTTM'] = pd.to_datetime(data['END_DTTM'], format='%d-%m-%Y %H:%M')

# Высчитываем difference
data['difference'] = data['END_DTTM'] - data['START_DTTM']
```

```python
data
```

```python
data = data.loc[data.difference > '5m']
data['SUBS_ID'].sort_values().values
```

```python
data = data[['ACT_DTTM', 'FILIAL_ID', 'difference']]
data
```

```python
data.to_csv('final_data.csv', sep=';')
```
