```python
taxi_count_borough = taxi['borough'].value_counts()
taxi_count_borough
```

```python
# Выяснить, общее число поездок за весь период
```

```python
taxi_sum_pickups = taxi.aggregate({'pickups' : 'sum'})
taxi_sum_pickups
```

```python
# Выяснить, из какого района было совершено наибольшее количество поездок за весь период
```

```python
taxi_sum_pickups_by_borough = taxi.groupby(['borough']).aggregate({'pickups' : 'sum'})
taxi_sum_pickups_by_borough
```

```python
# Выяснить, из какого района было совершено наименьшее количество поездок за весь период через функцию idxmin
```

```python
min_pickups = taxi.groupby(['borough']).aggregate({'pickups' : 'sum'})
min_pickups.idxmin()
```

```python
# Районы, из которых по праздникам в среднем поступает больше заказов, чем в обычные дни
```

```python
taxi_holiday = taxi.groupby(['borough', 'hday']) \
    .aggregate({'pickups' : 'mean'}) \
    .sort_values(['pickups'], ascending = False)
taxi_holiday
```

```python
# Для каждого района посчитайте число поездок по месяцам. Отсортируйте полученные значения по убыванию и сохраните
```

```python
pickups_by_mon_bor = taxi.groupby(['pickup_month', 'borough'], as_index=False) \
    .aggregate({'pickups' : 'sum'}) \
    .sort_values(['pickups'], ascending = False)
pickups_by_mon_bor
```

```python
# Напишите функцию temp_to_celcius, которая получает на вход колонку с температурой в °F, 
# и возвращает значения, переведенные в градусы Цельсия.
```

```python
def temp_to_celcius(temp):
    celc = (temp - 32) * 5.0 / 9.0
    return celc

temp = taxi.temp
print(temp_to_celcius(taxi.temp))
```
